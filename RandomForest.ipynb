## Attempt 2: Dropping some features using chi2

```python
# Try removing the Publishers
publishers = [p for p in X.columns.tolist() if p.startswith("Publisher_")]
```

```python
chi2_selector = SelectKBest(chi2, k='all')
chi2_selector.fit(X, y)
pvals = pd.DataFrame(chi2_selector.pvalues_, index=chi2_selector.feature_names_in_, columns=['p-value'])

filtered_features = pvals[pvals['p-value'] < 0.01].index.tolist()
filtered_features = [f for f in filtered_features]
print(len(filtered_features), " features selected.")
filtered_features
```

```python
X_filtered = X[filtered_features]
```

```python
rf.fit(X_filtered, y)
cross_val_report(rf, X_filtered, y)[1]
# Increased all
```
