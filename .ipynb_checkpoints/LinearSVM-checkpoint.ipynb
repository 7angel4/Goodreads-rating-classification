## Chi-squared selection

```python
chi2_selector = SelectKBest(chi2, k='all')
chi2_selector.fit(X_filtered, y_train)
pvals = pd.DataFrame(chi2_selector.pvalues_, index=chi2_selector.feature_names_in_, columns=['p-value'])
filtered_features = pvals[pvals['p-value'] < 0.05].index.tolist()
print(len(filtered_features), "multinomial features are significant.")
filtered_features
```

```python
X_filtered2 = X_filtered[filtered_features]
```

```python
cross_val_report(svm, X_filtered2, y_train)
```

## RFECV

## Visualize data

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(X_train.values)
df = train_df_ohe

df['pca-one'] = pca_result[:,0]
df['pca-two'] = pca_result[:,1] 
df['pca-three'] = pca_result[:,2]

print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
df
```

```python
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="pca-one", y="pca-two",
    hue="rating_label",
    palette=sns.color_palette("hls", 10),
    data=train_df_ohe,
    legend="full",
    alpha=0.3
)
```

```python
ax = plt.figure(figsize=(16,10)).gca(pro)
ax.scatter(
    xs=df["pca-one"], 
    ys=df["pca-two"], 
    zs=df["pca-three"], 
    c=y_train, 
    cmap='tab10'
)
ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')
plt.show()
```
