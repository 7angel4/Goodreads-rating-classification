# Stacking

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, CategoricalNB, ComplementNB
from mlxtend.classifier import StackingCVClassifier
from mlxtend.feature_selection import ColumnSelector
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression

cnb = make_pipeline(ColumnSelector(filtered_features),
                      ComplementNB())
mnb = make_pipeline(ColumnSelector(filtered_features),
                      MultinomialNB(alpha=0.1))

cvsnb = StackingCVClassifier(classifiers=[cnb, mnb], 
                            meta_classifier=LogisticRegression(),
                            random_state=30027)

cvsnb.fit(X_filtered2, y)
```

```python
cross_val_report(cvsnb, X_filtered2, y)
```

```python
param_grid = {'pipeline-1__complementnb__alpha': [i/100 for i in range(1,42,5)], 
              'pipeline-2__multinomialnb__alpha': [i/100 for i in range(1,42,5)]}

grid = tune_hyperparameter(cvsnb, param_grid, X_filtered2, y, cv=2)
grid
```

```python
grid.loc[grid['rank_test_accuracy']==1]['params'].values
```

```python
cnb = make_pipeline(ColumnSelector(filtered_features),
                      ComplementNB(alpha=0.11))
mnb = make_pipeline(ColumnSelector(filtered_features),
                      MultinomialNB(alpha=0.41))

cvsnb = StackingCVClassifier(classifiers=[cnb, mnb], 
                            meta_classifier=LogisticRegression(),
                            random_state=30027)

cvsnb.fit(X_filtered2, y)
cross_val_report(cvsnb, X_filtered2, y)
```
